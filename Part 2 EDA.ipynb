```python
# Get data

df = pd.read_csv('C:\\Users\\javwyk\OneDrive - MMI HOLDINGS LTD\\GitHub\Github-practice-assignment\\data\\tmdb_results_combined.csv.gz', low_memory = False)
```

```python
# Exclude movies without zeroes for revenue and budget; count the remaining movies

df_some_valid = df[(df['budget'] > 0) | (df['revenue'] > 0)]
len(df_some_valid)
```

```python
# Find the distribution of movies across certification categories

df_some_valid['certification'].value_counts()
```

```python
# Find the average budget for movies in each certification category

df_budget_summary = df_some_valid[df['budget'] > 0].groupby('certification')[['budget']].mean().round()
df_budget_summary
```

```python
# Find the average revenue for movies in each certification category

df_revenue_summary = df_some_valid[df['revenue'] > 0].groupby('certification')[['revenue']].mean().round()
df_revenue_summary
```
